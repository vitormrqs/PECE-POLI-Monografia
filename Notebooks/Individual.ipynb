## Funcoes

```python
#Conversor em SHA1
def convert_to_sha1(name):
    sha1 = hashlib.sha1(name.encode())
    return sha1.hexdigest()

#CPF fake
def generate_random_cpf():
    cpf = [random.randint(0, 9) for _ in range(9)]  

    # Cálculo do primeiro dígito verificador
    soma = sum((i + 1) * v for i, v in enumerate(cpf))
    d1 = (soma * 10) % 11
    cpf.append(d1 if d1 < 10 else 0)

    # Cálculo do segundo dígito verificador
    soma = sum((i + 2) * v for i, v in enumerate(cpf))
    d2 = (soma * 10) % 11
    cpf.append(d2 if d2 < 10 else 0)

    return f"{cpf[0]}{cpf[1]}{cpf[2]}{cpf[3]}{cpf[4]}{cpf[5]}{cpf[6]}{cpf[7]}{cpf[8]}{cpf[9]}{cpf[10]}"

#Idade fake
def generate_birth_date():
    year = random.randint(1950, 2003)  
    month = random.randint(1, 12)  
    day = random.randint(1, 28)  

    return f"{year}{month:02d}{day:02d}"  

#Formata nomes
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([char for char in nfkd_form if not unicodedata.combining(char)])

def format_names(input_list):
    return [remove_accents(name).upper() for name in input_list]
```

```python
names_list = format_names(names_list)
random_cpf_list = [generate_random_cpf() for _ in names_list]
hashes = [convert_to_sha1(name) for name in names_list]
birth_dates = [generate_birth_date() for age in names_list]

for name, hash_value, cpf, age in zip(names_list, hashes, random_cpf_list, birth_dates):
        print((
        f'{{'
        f'"COD_CHAV_CORO_CLIE": "{hash_value}",\n'
        f'"NOM_IDEF_PESS": "{name}",\n'
        f'"NUM_CPF_CNPJ": "{cpf}",\n'
        f'"COD_TIPO_PESS": "F",\n'
        f'"DAT_NASC": "{age}"'
        f'}},'
    ))
```
